```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
tags = pd.read_csv("Tags.csv", index_col=0)
```

```python
questions = pd.read_csv("Questions_new.csv", index_col=0, parse_dates=["CreationDate"])
questions["Tags"] = [[]] * len(questions)
for index, body in tags["Tag"].to_dict().items():
    questions.at[index, "Tags"] = questions.at[index, "Tags"] + [body]
```

```python
answers = pd.read_csv("Answers_new.csv", index_col=0, parse_dates=["CreationDate"])
```

```python
questions
```

```python
answers
```

```python
questions["Answers"] = 0
answer_question_ids = answers["ParentId"].to_dict().values()
for i in answer_question_ids:
    questions.at[i, "Answers"] += 1
```

```python
import numpy as np
max_len = np.percentile(questions["Length"], 90) 

questions_restricted = questions[questions["Length"] <= max_len]
#question_lengths = pd.cut(questions_restricted["Length"], 50)
#graph1 = questions_restricted["Answers"].groupby(question_lengths).mean().plot.line(xlabel="Question Length", ylabel="Mean #Answers")

bins = 50

_, edges = pd.cut(questions_restricted["Length"], bins, retbins=True)
labels = [f'{abs((edges[i] + edges[i+1])/2):.0f}' for i in range(len(edges)-1)]

question_lengths = pd.cut(questions_restricted["Length"], bins, labels=labels)
graph1 = questions_restricted["Answers"].groupby(question_lengths).mean().plot.line(xlabel="Question Length", ylabel="Mean #Answers")
```
