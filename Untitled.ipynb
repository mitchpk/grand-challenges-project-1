```python
import numpy as np
max_len = np.percentile(questions["Length"], 90) 

questions_restricted = questions[questions["Length"] <= max_len]
#question_lengths = pd.cut(questions_restricted["Length"], 50)
#graph1 = questions_restricted["Answers"].groupby(question_lengths).mean().plot.line(xlabel="Question Length", ylabel="Mean #Answers")

bins = 50

_, edges = pd.cut(questions_restricted["Length"], bins, retbins=True)
labels = [f'{abs((edges[i] + edges[i+1])/2):.0f}' for i in range(len(edges)-1)]

question_lengths = pd.cut(questions_restricted["Length"], bins, labels=labels)
graph1 = questions_restricted["Answers"].groupby(question_lengths).mean().plot.line(xlabel="Question Length", ylabel="Mean #Answers")
```

```python
test = questions[questions["Answers"] > 0]
test2 = test.resample(rule="M", on="CreationDate")["Tags"].sum().explode().groupby("CreationDate").value_counts().reset_index(name="Count").set_index("CreationDate")
test2
```

```python
# Tags used by answered questions
test3 = test2.loc['2016-01-01':'2016-08-01'].pivot(columns="Tags", values="Count").fillna(0)
test3[test3.sum().sort_values(ascending=False).head(10).index].plot.line(xlabel="Date", ylabel="Answered questions per month")
```

```python
test4 = questions[questions["Answers"] == 0]
test5 = test4.resample(rule="M", on="CreationDate")["Tags"].sum().explode().groupby("CreationDate").value_counts().reset_index(name="Count").set_index("CreationDate")
test5
```

```python
# Tags used by unanswered questions
test6 = test5.loc['2016-01-01':'2016-08-01'].pivot(columns="Tags", values="Count").fillna(0)
test6[test6.sum().sort_values(ascending=False).head(10).index].plot.line(xlabel="Date", ylabel="Unanswered questions per month")
```
